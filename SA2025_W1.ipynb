### Design a multi-line plot using Matplotlib or Seaborn that shows the evolution of average mpg over the years, separately for each origin

```python


# Clean column names
df.columns = df.columns.str.strip().str.lower()

# Rename model_year to year if needed
if 'model_year' in df.columns and 'year' not in df.columns:
    df.rename(columns={'model_year': 'year'}, inplace=True)

# Check if required columns exist
required_columns = {'year', 'origin', 'mpg'}
if not required_columns.issubset(df.columns):
    print(f"Error: Missing columns {required_columns - set(df.columns)} in DataFrame")
else:
    # Convert 'year' to numeric type
    df['year'] = pd.to_numeric(df['year'], errors='coerce')
    
    # Drop rows with missing values
    df = df.dropna(subset=['year', 'origin', 'mpg'])

    # Group by year and origin, then calculate average MPG
    mpg_trend = df.groupby(['year', 'origin'])['mpg'].mean().reset_index()

    # Create multi-line plot using Seaborn
    plt.figure(figsize=(10, 6))
    sns.lineplot(data=mpg_trend, x="year", y="mpg", hue="origin", marker="o")

    # Formatting
    plt.xlabel("Year")
    plt.ylabel("Average MPG")
    plt.title("Evolution of Average MPG Over the Years by Origin")
    plt.legend(title="Origin")
    plt.grid(True)

    # Show plot
    plt.show()
```

### We define a “consistent” car model as one that was produced over multiple years and had very low variation in mpg across those years (standard deviation < 1.0).

Tasks:

Identify car names that appear in more than one model_year.

For each such name, compute the standard deviation of mpg across years.

Return the car(s) with the lowest variation in mpg, among those with at least 2 appearances and std(mpg) < 1.0.

Report the model name(s), number of appearances, and the average mpg.

Bonus:
Sort the result by number of appearances (descending), then mpg (descending).

```python
import pandas as pd

# Ensure column names are clean
df.columns = df.columns.str.strip().str.lower()

# Check if required columns exist
required_columns = {'name', 'year', 'mpg'}
if not required_columns.issubset(df.columns):
    print(f"Error: Missing columns {required_columns - set(df.columns)} in DataFrame")
else:
    # Compute car appearance counts
    car_counts = df.groupby('name')['year'].count()

    # Filter cars appearing in multiple model years
    consistent_cars = car_counts[car_counts > 1].index

    # Compute standard deviation of MPG for each car
    mpg_std = df[df['name'].isin(consistent_cars)].groupby('name')['mpg'].std()

    # Filter cars with std(mpg) < 1.0
    low_variation_cars = mpg_std[mpg_std < 1.0].index

    # Compute average MPG and appearances
    mpg_avg = df[df['name'].isin(low_variation_cars)].groupby('name')['mpg'].mean()
    appearances = df[df['name'].isin(low_variation_cars)].groupby('name')['year'].count()

    # Combine results into DataFrame
    result = pd.DataFrame({'Appearances': appearances, 'Avg MPG': mpg_avg, 'MPG Std': mpg_std[low_variation_cars]})

    # Bonus: Sort by appearances (descending), then avg MPG (descending)
    result = result.sort_values(by=['Appearances', 'Avg MPG'], ascending=[False, False])

    # Display final result
    print(result)
```
